```python
%pip install pandas
%pip install matplotlib
%pip install seaborn
%pip install numpy
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.describe(include='all')
```

```python
# check null 

df.isnull().sum()
```

```python
df[df['Review Rating'].isna()].head()
```

```python
# Handling Missingh Values
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
# Changing column names to snake_case

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df= df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
'promo_code_used' in df.columns
```

```python
# categorical mapping
age_labels = ['young_adult', 'adult', 'middle_aged', 'senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=age_labels)

df[['age', 'age_group']]
# df.columns
'promo_code_used' in df.columns
```

```python
frequency_mapping = {
'Fortnightly': 14,
'Weekly': 7,
'Monthly':30,
'Quarterly' : 90,
'Bi-weekly': 14,
'Annually': 365,
'Every 3 Months': 90
}

df['purchase_frequency'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency','frequency_of_purchases']].sample(7)
```
